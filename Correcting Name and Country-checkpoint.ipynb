```python
import csv
import cv2
import pytesseract
import re
import numpy as np
import matplotlib.pyplot as plt




def read_text_from_video_frame(video_file, text_roi):
    # Read the video file
    cap = cv2.VideoCapture(video_file)

    # Read the first frame of the video
    for x in range(1,25):
        ret, frame = cap.read()
    roi = frame[325:365,210:600]
    roi_map = frame[320:365,130:200]
    plt.imshow(cv2.cvtColor(roi_map, cv2.COLOR_BGR2RGB))
    #cv2.imshow("frame", frame)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()

    # Convert the ROI to grayscale
    gray = cv2.cvtColor(roi, cv2.COLOR_BGR2GRAY)
    gray_map = cv2.cvtColor(roi_map, cv2.COLOR_BGR2GRAY)
    # Apply thresholding to the grayscale image
    threshold = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]
    threshold_map = cv2.threshold(gray_map, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Get the text from the thresholded image using Tesseract OCR
    text = pytesseract.image_to_string(threshold)
    text_map=pytesseract.image_to_string(threshold_map)
    # Release the video capture object and close all windows
    cap.release()
    #cv2.destroyAllWindows()
    name = text.split('\n')[0]

    country = text_map.split('\n')[0]
    return name,country



with open('scores.csv', 'r') as file:
    # Create a reader object
    reader = csv.reader(file)

    # Skip the header row
    next(reader)

    # Loop through each row and print the name column
    for x in range(1, 93):
        competition = "Men's 10m Platform Semi-Final - Diving ï½œ Tokyo 2020 Replays.mp4"
        fileName = str(x)+"_"+competition
        text_roi = (100, 200, 500, 400)
        video_file="ExtractedVideos/"+fileName
        name,country = read_text_from_video_frame(video_file, text_roi)
        print(str(x)+ "|"+name+"|"+country)
        #print(str(x)+" "+country)
```
