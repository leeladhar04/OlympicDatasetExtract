```python
def is_dive_started(img_rgb,template):

    desired_width = 1625
    desired_height = 911
    img_rgb = cv2.resize(img_rgb, (desired_width,desired_height))

    assert img_rgb is not None, "file could not be read, check with os.path.exists()"
    img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
    #plt.imshow(cv.cvtColor(template, cv.COLOR_BGR2RGB))

    assert template is not None, "file could not be read, check with os.path.exists()"
    w, h = template.shape[::-1]
    res = cv2.matchTemplate(img_gray,template,cv2.TM_CCOEFF_NORMED)
    threshold = 0.3
    loc = np.where( res >= threshold)


    cropped_image=img_rgb.copy()

    for pt in zip(*loc[::-1]):
        x, y = pt
        w, h = template.shape[::-1]
        # Crop the matched region
        cropped_image = img_rgb[y:y+h, x:x+w]

        # Save the cropped image
        #cv.imwrite('cropped_image.jpg', cropped_image)

        # Draw a rectangle around the matched region
        #cv.rectangle(img_rgb, pt, (x + w, y + h), (0, 0, 255), 2)
    return len(loc[0]) > 0
```

```python
def extract_clip(input_video_path, output_video_path, start_time, end_time):
    #print(input_video_path)
    #print(output_video_path)
    #print(start_time)
    #print(end_time)
    # Create a VideoFileClip object for the input video
    video_clip = VideoFileClip(input_video_path)

    # Extract the subclip between start_time and end_time
    clip = video_clip.subclip(start_time, end_time)

    # Write the clip to the output video file
    clip.write_videofile(output_video_path)

    # Close the VideoFileClip object
    video_clip.close()
```

```python
def GetPoseData(filename,template):
    ended=False
    dive_started=False
    frames_passed=0;
    empty_frames_passed=0
    dive_start_frame=0;
    mp_pose = mp.solutions.pose
    mp_drawing = mp.solutions.drawing_utils
    pose = mp_pose.Pose(static_image_mode=False, min_detection_confidence=0.5, min_tracking_confidence=0.5,model_complexity=2)
    points = mp_pose.PoseLandmark
    data = []
    data.append("Frame")
    for p in points:
        x = str(p)[13:]
        #x = p.name()
        data.append(x + "_x")
        data.append(x + "_y")
        data.append(x + "_z")
        data.append(x + "_vis")
    data = pd.DataFrame(columns = data)

    #print(data)
    filepath="ExtractedVideos/"+filename+".mp4"
    cap = cv2.VideoCapture(filepath)
    frameNo=0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frameNo=frameNo+1
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        if(not dive_started):
            b=is_dive_started(frame,template)
            if(b):
                dive_started=True
                dive_start_frame=frameNo
                print("dive started")
                print("Start time "+str(frameNo/25))

        if(not dive_started):
            continue


        if(empty_frames_passed>50 or frames_passed>500):
            ended=True
            print("End time "+str(frameNo/25))
            break

        results = pose.process(frame)
        info=(frameNo,)
        mp_drawing = mp.solutions.drawing_utils
        mp_drawing.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

        if results.pose_landmarks:
            for landmark in results.pose_landmarks.landmark:
                info +=(landmark.x, landmark.y, landmark.z) + tuple([landmark.visibility])
            data.loc[len(data.index)]=info
            empty_frames_passed=0

        else:
            empty_frames_passed+=1



        frames_passed+=1

    #print(data)

    data.to_csv("ExtractedPoseData/"+filename+".csv")
    if ended:
        extract_clip(filepath,'ExtractedPoseVideos/'+filename+".mp4",dive_start_frame/25,frameNo/25)
    cap.release()
    cv2.destroyAllWindows()


```

```python
csv_file = 'scores.csv'
df = pd.read_csv(csv_file)
template = cv2.imread('HelperImages/template_oympics.png',cv2.IMREAD_GRAYSCALE)
for index, row in df.iterrows():
    matchName=row['Match name']
    sno=row['Sno']
    isFile = os.path.isfile('ExtractedPoseData/'+str(sno)+"_"+matchName+".csv")

    isFileVid = os.path.isfile('ExtractedPoseVideos/'+str(sno)+"_"+matchName+".mp4")
    if(isFile or not isFileVid):
        pose_df = pd.read_csv('ExtractedPoseData/'+str(sno)+"_"+matchName+".csv")
        if pose_df.shape[0]>50:
            continue
        GetPoseData(str(sno)+"_"+matchName,template)
#frame = cv2.imread('HelperImages/mainRio.png')
#print(dive_started(frame,template))

print("Completed")
```
