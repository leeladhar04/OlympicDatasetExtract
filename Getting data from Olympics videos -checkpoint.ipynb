```python
import csv
import cv2
from pytesseract import image_to_string
import pandas as pd
import re
import os
from moviepy.video.io.VideoFileClip import VideoFileClip

def extract_clip(input_video_path, output_video_path, start_time, end_time):
    print(input_video_path)
    print(output_video_path)
    print(start_time)
    print(end_time)
    # Create a VideoFileClip object for the input video
    video_clip = VideoFileClip(input_video_path)

    # Extract the subclip between start_time and end_time
    clip = video_clip.subclip(start_time, end_time)

    # Write the clip to the output video file
    clip.write_videofile(output_video_path)

    # Close the VideoFileClip object
    video_clip.close()

def tryextract_judges_scores(text):
    sequences = re.findall(r'\d+(?:\.\d+)?(?:\s+\d+(?:\.\d+)?)*', text,re.IGNORECASE)
    longest_sequence = max(sequences, key=lambda seq: len(seq.split()))
    numbers = [float(num) if '.' in num else int(num) for num in longest_sequence.split()]
    numbers = [num / 10 if num >= 10 else num for num in numbers]
    scores = ', '.join(str(n) for n in numbers)
    return scores

def tryextract_final_score(text):
    finalScore=""
    pattern1 = r"round \d+ (\d+\.\d+)"
    match1 = re.search(pattern1, text,re.IGNORECASE)
    pattern2=r'\d{2}\.\d{2}'
    match2 = re.search(pattern2, text,re.IGNORECASE)
    if match1:
        finalScore = match1.group(1)
    elif match2:
        finalScore = match2.group(0)
    return finalScore

def tryextract_total_score(text):
    totalScore=""
    pattern1 = r"total \ (\d+\.\d+)"
    match1 = re.search(pattern1, text,re.IGNORECASE)
    pattern2=r'\d{2}\.\d{2}'
    match2 = re.search(pattern2, text,re.IGNORECASE)
    if match1:
        totalScore = match1.group(1)
    elif match2:
        totalScore = match2.group(0)
    return totalScore

def extract_somersaults(text):
    match1 = re.search(r'\d+% SOMERSAULTS', text,re.IGNORECASE)
    if match1:
        somersaults = float(match1.group(0).split('%')[0])+0.5
        return somersaults
    match2 = re.search(r'(\d+(\.\d+)?)\D*\b(somersaults?)\b', text, re.IGNORECASE)
    if match2:
        somersaults = match2.group(1)
        return somersaults

def extract_twists(text):
    match1 = re.search(r'\d+% TWISTS', text,re.IGNORECASE)
    if match1:
        twists = float(match1.group(0).split('%')[0])+0.5
        return twists
    match2 = re.search(r'(\d+(\.\d+)?)\D*\b(twists?)\b', text, re.IGNORECASE)
    if match2:
        twists = match2.group(1)
        return twists

def extract_diveinfo(text):
    country=""
    firstName=""
    secondName=""
    difficulty = ""
    round = ""
    diveGroup=""
    divePosition=""
    somersaults=""
    twists=""
    country = text.split()[0]
    firstName=text.split()[1]
    secondName=text.split()[2]
    somersaults = extract_somersaults(text)
    twists = extract_twists(text)
    match = re.search(r'DIFFICULTY (\d+(\.\d+)?)', text, re.IGNORECASE)
    if match:
        difficulty = match.group(1)
    match = re.search(r'ROUND (\d+)', text, re.IGNORECASE)
    if match:
        round = match.group(1)
    match = re.search(r'(\w+) POSITION', text,re.IGNORECASE)
    if match:
        divePosition = match.group(1)
    match = re.search(r'(ARMSTAND|BACK|INWARD|FORWARD|REVERSE|TWISTER|ARMSTAND BACK|ARMSTAND FORWARD|BACK ARMSTAND|BACK FORWARD|BACK INWARD|BACK REVERSE|BACK TWISTER|FORWARD ARMSTAND|FORWARD BACK|FORWARD INWARD|FORWARD REVERSE|FORWARD TWISTER|INWARD BACK|INWARD FORWARD|INWARD REVERSE|INWARD TWISTER|REVERSE BACK|REVERSE FORWARD|REVERSE INWARD|REVERSE TWISTER|TWISTER BACK|TWISTER FORWARD|TWISTER INWARD|TWISTER REVERSE)', text, re.IGNORECASE)
    if match:
        diveGroup = match.group(1)


    return round,country,firstName,secondName,difficulty,divePosition,somersaults,diveGroup,twists

def extract_score(text):
    # Initialize variables to store the score, round, penalty and difficulty
    finalScore = ""
    penalty = ""
    judgesScore=""
    judgesScore=tryextract_judges_scores(text)
    finalScore=tryextract_final_score(text)
    totalScore=tryextract_total_score(text)
    match = re.search(r'PENALTY (\d+)', text, re.IGNORECASE)
    if match:
        penalty = match.group(1)
    return finalScore,totalScore,penalty,judgesScore

def WriteDiveinfoInCSV(fileName,info):
    headers=["round","country","firstName","secondName","difficulty","dive"]
    with open(fileName, "r") as f:
        reader = csv.reader(f)
        first_row = next(reader, None)
        if first_row != headers:
            # Open the file in write mode to add headers
            with open(fileName, "w", newline="") as f:
                writer = csv.writer(f)
                writer.writerow(headers)
                # Write the scores to the CSV file
                writer.writerow(score)
        else:
            # Open the file in append mode and write scores
            with open(fileName, "a", newline="") as f:
                writer = csv.writer(f)
                writer.writerow(score)


def WriteInCSV(fileName,score):
    headers = ["Video Name","Sno","Start Time","Round","Country","First Name","Last Name","Difficulty","Dive Position","Somersaults","Dive Group","Twist","Final Score","Total Score","Penalty","Judges Score","End Time"]
    # Open the file in read mode to check headers
    if os.path.exists(fileName):
    # File exists, open in read mode
        with open(fileName, "r") as f:
            reader = csv.reader(f)
            first_row = next(reader, None)
            if first_row != headers:
                # Open the file in write mode to add headers
                with open(fileName, 'w', encoding='utf-8', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow(headers)
                    # Write the scores to the CSV file
                    writer.writerow(score)
            else:
            # Open the file in append mode and write scores
                with open(fileName, 'a', encoding='utf-8', newline='') as f:
                    writer = csv.writer(f)
                    writer.writerow(score)
    else:
    # File does not exist, open in write mode and write headers
        with open(fileName, 'w', encoding='utf-8', newline='') as f:
            writer = csv.writer(f)
            writer.writerow(headers)
            # Write the scores to the CSV file
            writer.writerow(score)


# Open the video using OpenCV
filePath="DownloadedVideos/Women's 3m Springboard Semi-Final - Diving ｜ Tokyo 2020 Replays [850af038-fca9-4365-ab54-c4f6e14499d6].mp4"
videoName="Women's 3m Springboard Semi-Final - Diving ｜ Tokyo 2020 Replays"
cap = cv2.VideoCapture(filePath)
fps=cap.get(cv2.CAP_PROP_FPS)
print(fps)
minToSkip=9
framesToSkip=fps*60*minToSkip
framesToSkipEverySec=fps-1
currentFrame=0
currentFrameInSec=0
diveStarted=False
sno=1
row=[]

# Initialize a list to store the scores

print("start")

oldText="old"
# Process each frame of the video
while True:
    # Read the next frame
    ret, frame = cap.read()
    currentFrame+=1
    currentFrameInSec+=1
    if currentFrame<framesToSkip :
        if currentFrame%(fps*60)==0:
            print(currentFrame//(fps*60) ," min(s) Skipped")
        continue

    if currentFrameInSec<framesToSkipEverySec:
        continue



    # Break the loop if the video has ended
    if not ret:
        print("completed")
        break

    current_pos = cap.get(cv2.CAP_PROP_POS_MSEC)
    total_seconds = int(current_pos // 1000)
    minutes = total_seconds // 60
    seconds = total_seconds % 60
    timeInVideo =f"{minutes} minutes and {seconds} seconds"
    # Convert the frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Use OCR to extract the text from the frame
    text = image_to_string(gray,lang="eng")

    words = text.lower().split()  # split the text into a list of words
    filtered_words = [word for word in words if (word != 'tokyo' and word!='2020' and word!='202' and word!='toky' and word!='tokyo2020' and word!="0ky0" and word!="toky0") ]  # use list comprehension to remove the word "hello"
    text = ' '.join(filtered_words)  # join the remaining words back into a string

    if not diveStarted and "round" in text.lower() and "difficulty" in text.lower() and "penalty" not in text.lower() and "position" in text.lower():
        diveStarted=True
        diveinfo = extract_diveinfo(text)
        row=diveinfo
        row=(total_seconds,)+row
        row=(sno,)+row
        row=(videoName,)+row
        print(row)
        print(timeInVideo," dive started")
        #WriteDiveinfoInCSV("diveinfo.csv",score)
    #if diveStarted and text!="":
     #   print(text)
    if diveStarted and( ("difficulty" in text.lower() and "penalty" in text.lower()) | ("penalty" in text.lower()) ):
        print(timeInVideo," dive scores announced")
        score=extract_score(text)
        row=row+score
        row+=(total_seconds,)
        WriteInCSV("scores.csv",row)
        extract_clip(filePath,'ExtractedVideos/'+str(row[1])+'_'+videoName+'.mp4',int(row[2]),int(total_seconds))
        diveStarted=False
        sno+=1
        print(row)

    #cv2.imshow("frame", frame)
    #cv2.waitKey(0)
    #cv2.destroyAllWindows()
    #cv2.imwrite("frame.jpg", frame)

    currentFrameInSec=0

# Release the video capture
cap.release()
```
